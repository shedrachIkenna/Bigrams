```python
import torch 
```

```python
xs = [
  [2.0, 3.0, -1.0], # input 1 => the desired target is 1.0 
  [3.0, -1.0, 0.5], # input 2 => the desired target is -1.0 
  [0.5, 1.0, 1.0],  # input 3 => the desired target is -1.0 
  [1.0, 1.0, -1.0], # input 4 => the desired taregt is 1.0 
]
ys = [1.0, -1.0, -1.0, 1.0]
```

```python
xs = torch.tensor(xs)
ys = torch.tensor(ys)
```

```python
# xs @ w 
# Lets just create a one-layer NN 

# xs = (4, 3) matrix 
# w = (3, 1) matrix 

# xs @ w = (4, 1)

# create a generator to that we will have identical seed generation for the random weights 
g = torch.Generator().manual_seed(2147483547)
W = torch.randn((3, 1), generator=g, requires_grad=True)
```

```python
for _ in range(20):
    ypred = xs @ W
    loss = ((ypred.squeeze() - ys)**2).sum()
    loss.backward()
    W.data -= 0.05 * W.grad
    W.grad.zero_()
    print(loss.item())
```

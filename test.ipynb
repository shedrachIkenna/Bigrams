```python
import torch 
```

```python
xs = [
  [2.0, 3.0, -1.0], # input 1 => the desired target is 1.0 
  [3.0, -1.0, 0.5], # input 2 => the desired target is -1.0 
  [0.5, 1.0, 1.0],  # input 3 => the desired target is -1.0 
  [1.0, 1.0, -1.0], # input 4 => the desired taregt is 1.0 
]
ys = [1.0, -1.0, -1.0, 1.0]
```

```python
# first we need to convert these python lists to tensors 
xs = torch.tensor(xs)
ys = torch.tensor(ys)
```

```python
# compute wx + b (for simplicity, we are going to leave out the bias term b )
# Normally its x @ w 

# initialize random weights from a normal distribution 
# xs is a (4, 3) matrix 
# W should be a (3, 1) for a one layer NN 
# output matrix is (4, 3) @ (3, 1) => (4, 1) 
# (4, 1) gives us the 4 predicted values from the NN 
g = torch.Generator().manual_seed(2147483547)
W = torch.randn((3, 1), generator=g, requires_grad=True)
```

```python
W
```

```python
# compute x @ w 
ypred = xs @ W 
ypred 
```

```python
loss = sum([(ypred - ys)**2 for ys, ypred in zip(ys, ypred)])
loss
```

```python
loss.backward()
```

```python
W.grad
```

```python
W.data -= 0.05 * W.grad
```

```python
W.grad.zero_()
```

```python
for _ in range(20):
    ypred = xs @ W 
    loss = sum([(ypred - ys)**2 for ys, ypred in zip(ys, ypred)])
    loss.backward()
    W.data -= 0.05 * W.grad
    W.grad.zero_()
```

```python
for _ in range(20):
    ypred = xs @ W
    loss = ((ypred.squeeze() - ys) ** 2).sum()
    
    loss.backward()
    
    with torch.no_grad():
        W.data -= 0.05 * W.grad
    
    W.grad.zero_()
    
    print(loss.item())
```
